```python
# 3.8 Seeing the World
# Think of at least five places in the world you'd like to visit.
# Store locations in a list
# Print in original order

TRAVEL = ['Laos', 'Korea', 'Lithuania', 'Libia', 'Uruguay', 'Alaska']

print(TRAVEL)

# use sorted() to print in alphabetical order
# show your list is still in its original order by printing it

print(sorted(TRAVEL))
print(TRAVEL)


# use sorted() to print your list in reverse-alphabetical order by printing again

print(sorted(TRAVEL, reverse = True))


# use reverse() to change the order of our list. Print the list to show that its order has changed

TRAVEL.reverse()
print(TRAVEL)

# use reverse() to change the order of your list again. Print to show its original order

TRAVEL.reverse()
print(TRAVEL)

# use sort() to change your list so it's stored in alphabetical order. print to show its order has changed.

TRAVEL.sort()
print(TRAVEL)

# use sort() to change your list so it's stored in reverse-alphabetical order. Print the list to show that its order has changed.

TRAVEL.sort(reverse = True)
print(TRAVEL)
```

```python
# 3.10
# Think of things you could store in a list. Write a program that creates a list containing these items and then uses each function introduced in this chapter at least once.

MENU = ['orange chicken', 'teriyaki chicken', 'general shos chicken']

# index(), title(), append(), insert(), del(), pop(), remove(), sort(), sorted(), reverse=True, reverse(), len()

print(MENU)

MENU.append('white rice')
print(MENU)

MENU.insert(3, 'brown rice')
print(MENU)

print(f"We are out of {MENU[2].title()}.")

del MENU[2]
print(MENU)

REPLACEMENT = ['beef stif fry', 'bejing beef', 'honey walnut shrimp', 'eggplant tofu']

MENU.pop(0)
MENU.remove('teriyaki chicken')

MENU_TWO = MENU + REPLACEMENT

print(f"Sorry, we are out of chicken. Here's our replacement menu with {len(MENU_TWO)} options: {MENU_TWO}")

MENU_TWO.sort()
print(MENU_TWO)

print(sorted(MENU_TWO, reverse = True))
```
